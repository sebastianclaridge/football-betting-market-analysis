# 02 - Modelling
## 1. Paths, imports & settings

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import hashlib
import joblib
from dataclasses import dataclass

from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.preprocessing import StandardScaler, label_binarize
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
from sklearn.metrics import log_loss

# Project paths
PROJECT_DIR = Path(__file__).resolve().parents[1] if '__file__' in globals() else Path.cwd().parents[0]
DATA_DIR = PROJECT_DIR / "data"
RAW_DATA_DIR = DATA_DIR / "raw"
PROCESSED_DATA_DIR = DATA_DIR / "processed"

OUTPUT_DIR = PROJECT_DIR / "outputs"
FIGURES_DIR = OUTPUT_DIR / "figures"
REPORTS_DIR = OUTPUT_DIR / "reports"

for p in [OUTPUT_DIR, FIGURES_DIR, REPORTS_DIR, PROCESSED_DATA_DIR]:
    p.mkdir(parents=True, exist_ok=True)

pd.set_option("display.max_columns", 200)
RANDOM_STATE = 42

print("RAW_DATA_DIR:", RAW_DATA_DIR.resolve())
print("OUTPUT_DIR:", OUTPUT_DIR.resolve())
```

## 2. Load data & target

```python
# Load E0.csv
E0_PATH = RAW_DATA_DIR / "E0.csv"
df = pd.read_csv("../data/raw/E0.csv")

# Dates and sorting
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')
df = df.dropna(subset=['Date', 'FTR']).sort_values('Date').reset_index(drop=True)

# 3-way target
df['y_3way'] = df['FTR'].astype(str).str.upper().str.strip()
df['y_3way'].value_counts(dropna=False)
```

## 3. Odds to implied probabilities

```python
# Choose odds source
ODDS_COLS = ['B365H','B365D','B365A']
df = df.dropna(subset=ODDS_COLS).copy()

# Implied and overround
for col in ODDS_COLS:
    df[f"imp_{col}"] = 1.0 / df[col]
df['overround'] = df[[f"imp_{c}" for c in ODDS_COLS]].sum(axis=1).clip(lower=1e-12)

for col in ODDS_COLS:
    df[f"p_{col}"] = df[f"imp_{col}"] / df['overround']

# Check
df[['p_B365H','p_B365D','p_B365A',]].head()
```

## 4. Features, split & class order

```python
# Features
feature_cols = [
    'p_B365H','p_B365D','p_B365A',
    'B365H','B365D','B365A',
    'overround'
]

X = df[feature_cols].copy()
y = df['y_3way'].copy()

# Single fixed class order
CLASS_ORDER = np.array(['H','D','A'])

# Time-based split: 70/15/15
n = len(df)
train_end = int(n * 0.70)
valid_end = int(n * 0.85)

X_train, y_train = X.iloc[:train_end], y.iloc[:train_end]
X_valid, y_valid = X.iloc[train_end:valid_end], y.iloc[train_end:valid_end]
X_test, y_test = X.iloc[valid_end:], y.iloc[valid_end:]

len(X_train), len(X_valid), len(X_test)
```

## 5. Model & CV search

```python
softmax_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
    ("clf", LogisticRegression(
        penalty="l2",
        solver="lbfgs",
        max_iter=1000,
        multi_class="multinomial",
        random_state=RANDOM_STATE
    ))
])

param_grid = {
    "clf__C": np.logspace(-3, 2, 10),
    "clf__class_weight": [None, "balanced"]
}

tscv = TimeSeriesSplit(n_splits=5)

search = GridSearchCV(
    estimator=softmax_pipe,
    param_grid=param_grid,
    scoring="neg_log_loss",
    cv=tscv,
    n_jobs=-1,
    refit=True
)

search.fit(X_train, y_train)
best_softmax = search.best_estimator_
search.best_params_, -search.best_score_
```

## 6. Calibration & helpers

```python
# Calibrate on validation
calibrated_ovr = CalibratedClassifierCV(
    estimator=best_softmax,
    method="sigmoid",
    cv="prefit"
)
calibrated_ovr.fit(X_valid, y_valid)

# Helpers
def reorder_probs(clf, proba, class_order):
    idx = [list(clf.classes_).index(c) for c in class_order]
    return proba[:, idx]

def brier_multiclass(y_true, proba, class_order):
    Y = label_binarize(y_true, classes=class_order)
    return float(np.mean(np.sum((proba - Y)**2, axis=1)))

FIGURES_CAL_DIR = FIGURES_DIR / "notebook02_calibration"
FIGURES_CAL_DIR.mkdir(parents=True, exist_ok=True)

def reliability_plot_ovr(y_true, proba, label, tag, bins=10):
    y_bin = (y_true == label).astype(int)
    frac_pos, mean_pred = calibration_curve(y_bin, proba, n_bins=bins, strategy='uniform')
    
    plt.figure(figsize=(5,5))
    plt.plot(mean_pred, frac_pos, marker='o', linewidth=1)
    plt.plot([0,1],[0,1],'--')
    plt.title(f"Reliability ({tag}) â€” {label}")
    plt.xlabel("Mean predicted probability")
    plt.ylabel("Fraction of positives")
    plt.grid(True)
    plt.tight_layout()

    out_path = FIGURES_CAL_DIR / f"reliability_{tag}_{label}.png"
    plt.savefig(out_path, dpi=200)
    plt.show()
    print("Saved:", out_path)

def ece_binary(y_true_bin, proba, n_bins=20):
    bins = np.linspace(0.0, 1.0, n_bins+1)
    ids = np.digitize(proba, bins) - 1
    ece = 0.0
    for b in range(n_bins):
        m = ids == b
        if m.any():
            conf = proba[m].mean()
            acc = y_true_bin[m].mean()
            ece += m.mean() * abs(acc - conf)
    return float(ece)
```

## 7. Validation metrics (raw vs calibrated)

```python
# Raw / calibrated probabilities
proba_valid_raw = reorder_probs(best_softmax.named_steps['clf'], best_softmax.predict_proba(X_valid), CLASS_ORDER)
proba_valid_cal = reorder_probs(calibrated_ovr, calibrated_ovr.predict_proba(X_valid), CLASS_ORDER)

# Log loss
valid_ll_raw = log_loss(y_valid, proba_valid_raw, labels=CLASS_ORDER)
valid_ll_cal = log_loss(y_valid, proba_valid_cal, labels=CLASS_ORDER)
valid_brier_raw = brier_multiclass(y_valid, proba_valid_raw, CLASS_ORDER)
valid_brier_cal = brier_multiclass(y_valid, proba_valid_cal, CLASS_ORDER)

valid_metrics_df = pd.DataFrame({
    "metric": ["log_loss", "brier_multiclass"],
    "raw": [valid_ll_raw, valid_brier_raw],
    "calibrated": [valid_ll_cal, valid_brier_cal]
})
display(valid_metrics_df)

# Reliability curves
for i, c in enumerate(CLASS_ORDER):
    reliability_plot_ovr(y_valid, proba_valid_raw[:, i], c, tag="RAW_valid", bins=10)
for i, c in enumerate(CLASS_ORDER):
    reliability_plot_ovr(y_valid, proba_valid_cal[:, i], c, tag="CAL_valid", bins=10)

# ECE per class and save to reports
ece_valid = {}
for i, c in enumerate(CLASS_ORDER):
    yb = (y_valid == c).astype(int).values
    ece_valid[(c, "raw")] = ece_binary(yb, proba_valid_raw[:, i])
    ece_valid[(c, "cal")] = ece_binary(yb, proba_valid_cal[:, i])

ece_valid_df = pd.Series(ece_valid).unstack().reset_index().rename(columns={'level_0':'class'})
display(ece_valid_df)

# Save validation metrics
valid_metrics_path = REPORTS_DIR / "nb02_valid_metrics.csv"
ece_valid_path     = REPORTS_DIR / "nb02_ece_per_class.csv"
valid_metrics_df.to_csv(valid_metrics_path, index=False)
ece_valid_df.to_csv(ece_valid_path, index=False)
print("Saved:", valid_metrics_path)
print("Saved:", ece_valid_path)
```

## 8. Test metrics (raw vs calibrated), reliability & ECE

```python
# Test probabilities
proba_test_raw = reorder_probs(best_softmax.named_steps['clf'], best_softmax.predict_proba(X_test), CLASS_ORDER)
proba_test_cal = reorder_probs(calibrated_ovr, calibrated_ovr.predict_proba(X_test), CLASS_ORDER)

# Test metrics
test_ll_raw = log_loss(y_test, proba_test_raw, labels=CLASS_ORDER)
test_ll_cal = log_loss(y_test, proba_test_cal, labels=CLASS_ORDER)
test_brier_raw = brier_multiclass(y_test, proba_test_raw, CLASS_ORDER)
test_brier_cal = brier_multiclass(y_test, proba_test_cal, CLASS_ORDER)

metrics_df = pd.DataFrame({
    "metric": ["log_loss", "brier_multiclass"],
    "raw": [test_ll_raw, test_brier_raw],
    "calibrated": [test_ll_cal, test_brier_cal]
})
display(metrics_df)


# Reliability test
for i, c in enumerate(CLASS_ORDER):
    reliability_plot_ovr(y_test, proba_test_raw[:, i], c, tag="RAW_test", bins=10)
for i, c in enumerate(CLASS_ORDER):
    reliability_plot_ovr(y_test, proba_test_cal[:, i], c, tag="CAL_test", bins=10)


# ECE test per class
ece = {}
for i, c in enumerate(CLASS_ORDER):
    yb = (y_test == c).astype(int).values
    ece[(c, "raw")] = ece_binary(yb, proba_test_raw[:, i])
    ece[(c, "cal")] = ece_binary(yb, proba_test_cal[:, i])   

ece_df = pd.Series(ece).unstack().reset_index().rename(columns={'level_0':'class'})
display(ece_df)

# Save metrics & ECE to reports
REPORTS_DIR.mkdir(parents=True, exist_ok=True)
metrics_path = REPORTS_DIR / "nb02_test_metrics.csv"
ece_path = REPORTS_DIR / "nb02_ece_per_class.csv"
metrics_df.to_csv(metrics_path, index=False)
ece_df.to_csv(ece_path, index=False)
print("Saved:", metrics_path)
print("Saved:", ece_path)
```

## 9. Build prediction frame for the test slice

```python
# Slice rows that correspond to X_test
n_train, n_valid = len(X_train), len(X_valid)
pred_df = df.iloc[n_train + n_valid : ].copy()

# Attach probs
pred_df['proba_H_raw'], pred_df['proba_D_raw'], pred_df['proba_A_raw'] = proba_test_raw.T
pred_df['proba_H_cal'], pred_df['proba_D_cal'], pred_df['proba_A_cal'] = proba_test_cal.T

# Rename market probs for clarity
rename_map = {'p_B365H':'market_p_H','p_B365D':'market_p_D','p_B365A':'market_p_A'}
for k,v in rename_map.items():
    if k in pred_df.columns:
        pred_df = pred_df.rename(columns={k:v})

cols_show = [c for c in [
    "Date","HomeTeam","AwayTeam","FTR",
    "B365H","B365D","B365A",
    "market_p_H","market_p_D","market_p_A",
    "proba_H_raw","proba_D_raw","proba_A_raw",
    "proba_H_cal","proba_D_cal","proba_A_cal"
] if c in pred_df.columns]
pred_df[cols_show].head(5)
```

## 10. Export artifacts (models & csvs in correct directories)

```python
# Helper: match_id
def make_match_id(row):
    key = f"{pd.to_datetime(row['Date']).date()}|{row['HomeTeam']}|{row['AwayTeam']}"
    return hashlib.md5(key.encode()).hexdigest()[:16]

# Save trained models
joblib.dump(best_softmax, OUTPUT_DIR / "softmax_odds_base.joblib")
joblib.dump(calibrated_ovr, OUTPUT_DIR / "softmax_odds_calibrated_sigmoid.joblib")


# Save TEST-SET predictions
pred_full_path = REPORTS_DIR / "predictions_test_calibrated_multiclass.csv"
pred_df.to_csv(pred_full_path, index=False)

print("Saved test-set predictions to:", pred_full_path)

# Export Calibrated probabilities for the entire dataset
# Predict calibrated probabilities for ALL rows of X
X_all = X.copy()
proba_all_raw = calibrated_ovr.predict_proba(X_all)

# Ensure correct class order (['H','D','A'])
classes = list(calibrated_ovr.classes_)  # e.g. ['A','D','H']
idx_H = classes.index('H')
idx_D = classes.index('D')
idx_A = classes.index('A')

# Attach match_id to full df
df_full = df.copy()
df_full['match_id'] = df_full.apply(make_match_id, axis=1)

# Attach calibrated probabilities in correct order
df_full['p_H'] = proba_all_raw[:, idx_H]
df_full['p_D'] = proba_all_raw[:, idx_D]
df_full['p_A'] = proba_all_raw[:, idx_A]

# Minimal file for NB3
model_probs = df_full[['match_id','p_H','p_D','p_A']].copy()

# Defensive renormalisation
row_sum = model_probs[['p_H','p_D','p_A']].sum(axis=1).replace(0, np.nan)
model_probs[['p_H','p_D','p_A']] = model_probs[['p_H','p_D','p_A']].div(row_sum, axis=0)

MODEL_PROBS_PATH = OUTPUT_DIR / "model_probabilities.csv"
model_probs.to_csv(MODEL_PROBS_PATH, index=False)
print("Saved full-season model probabilities to:", MODEL_PROBS_PATH)

# Tidy version with metadata

tidy = df_full.rename(columns={
    'Date':'date',
    'HomeTeam':'home_team',
    'AwayTeam':'away_team'
})

tidy = tidy[['match_id','date','home_team','away_team','p_H','p_D','p_A']].copy()

# Renormalise again
row_sum = tidy[['p_H','p_D','p_A']].sum(axis=1).replace(0, np.nan)
tidy[['p_H','p_D','p_A']] = tidy[['p_H','p_D','p_A']].div(row_sum, axis=0)

TIDY_PATH = PROCESSED_DATA_DIR / "model_probabilities__tidy.csv"
tidy.to_csv(TIDY_PATH, index=False)

print("Saved tidy full dataset to:", TIDY_PATH)

print("\nExports complete:")
print(" -", OUTPUT_DIR / "softmax_odds_base.joblib")
print(" -", OUTPUT_DIR / "softmax_odds_calibrated_sigmoid.joblib")
print(" -", pred_full_path, "(test-only)")
print(" -", MODEL_PROBS_PATH, "(ALL matches â† used by NB3)")
print(" -", TIDY_PATH)
```

## 11. QC

```python
# Quick QC on model_probabilities.csv
tmp = pd.read_csv(MODEL_PROBS_PATH)
assert set(tmp.columns) == {'match_id','p_H','p_D','p_A'}
assert np.allclose(tmp[['p_H','p_D','p_A']].sum(axis=1), 1.0, atol=1e-6)
print("QC OK â€” outputs/model_probabilities.csv is ready for Notebook 3.")
```
